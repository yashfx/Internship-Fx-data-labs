Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

Reading the dataset

```python
df = pd.read_csv(r"D:\Internship Fx data labs\Day 9\creditcard.csv")
```

Preprocessing the data

```python
fraud=df[df["Class"]==1]
not_fraud=df[df["Class"]==0]
```

```python
print(fraud)
print(not_fraud)
```

```python
fraud.Amount.describe()
```

```python
not_fraud.Amount.describe()
```

```python
not_fraud=not_fraud.sample(n=400)
not_fraud
```

```python
palanced_dataset=pd.concat([fraud,not_fraud],axis=0)
palanced_dataset
```

```python
palanced_dataset["Class"].value_counts()
```

Splitting the data

```python
x=palanced_dataset.drop("Class",axis='columns')
y=palanced_dataset["Class"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

Building the model

```python
logs_model=LogisticRegression()
logs_model.fit(x_train,y_train)
```

```python
y_predict=logs_model.predict(x_test)
```

Evaluating the model

```python
from sklearn.metrics import precision_score,recall_score, confusion_matrix , classification_report
accuracy=accuracy_score(y_test,y_predict)
print("Accuracy= ",int(accuracy*100),'%')
precision = precision_score(y_test, y_predict)
print("Precision= ",int(precision*100),"%")
recall = recall_score(y_test, y_predict)
print("Recall= ",int(recall*100),"%")
cm = confusion_matrix(y_test, y_predict)
print("Confusion Matrix:")
print(cm)
report = classification_report(y_test, y_predict)
print("Classification Report:")
print(report)
```
